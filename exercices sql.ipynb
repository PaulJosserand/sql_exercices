# Question 1: Quel est le délais moyen entre deux achat d'un même user ?

```python
q="""

--On calcule la moyenne entre chaque achat pour chaque user_id


SELECT *,
    AVG(time_between_orders) AS avg_time_between_orders
FROM
(
    -- on calcule le nombre de jours écoulés entre 2 achats
    SELECT
        user_id,
        product_id,
        event_name,
        date,
        CASE 
            WHEN LAG(event_name) OVER (PARTITION BY user_id ORDER BY date) = 'order' 
            THEN date - LAG(date) OVER (PARTITION BY user_id ORDER BY date)
            ELSE 0 
        END AS time_between_orders
    FROM
    (
        --On extrait les lignes de commande des user_id précédemment extraits (ceux avec au moins 2 order)
        
        SELECT c.*
        FROM 
        (
        --On extrait les user_id qui ont réalisé plus d'une commande
        
        SELECT user_id,
           SUM(CASE WHEN event_name='order' THEN 1 ELSE 0 END) AS nbr_order
        FROM df_caudz
        GROUP BY user_id
        HAVING nbr_order
        ORDER BY nbr_order DESC
        )o
        LEFT JOIN df_caudz c ON o.user_id = c.user_id WHERE event_name ='order'
    )
)
GROUP BY user_id
ORDER BY avg_time_between_orders DESC
"""
sqldf(q)
```

# Question 2: Evaluation du tunnel de déperdition

**Précisions**

Ici, nous voulons évaluer pour 100 'seen', combien de 'add to cart', pour 100 'add to cart' combien de 'order'

### A) Sans distinction par produit

```python
q="""
SELECT
    ROUND((nbr_order*100/nbr_seen), 2) AS nbr_order_for_100_seen,
    ROUND((nbr_atc*100/nbr_seen), 2) AS nbr_atc_for_100_seen,
    ROUND((nbr_order*100/nbr_atc), 2) AS nbr_order_for_100_atc

FROM (
    SELECT
        SUM(CASE WHEN event_name = 'seen' THEN 1 ELSE 0 END) AS nbr_seen,
        SUM(CASE WHEN event_name = 'add to cart' THEN 1 ELSE 0 END) AS nbr_atc,
        SUM(CASE WHEN event_name = 'order' THEN 1 ELSE 0 END) AS nbr_order
    FROM df_caudz
    )
"""
df_question_2A = sqldf(q)
df_question_2A
```

### B) Avec distinction par produit

```python
q="""
SELECT
    product_id,
    ROUND((nbr_order*100/nbr_seen), 2) AS nbr_order_for_100_seen,
    ROUND((nbr_atc*100/nbr_seen), 2) AS nbr_atc_for_100_seen,
    ROUND((nbr_order*100/nbr_atc), 2) AS nbr_order_for_100_atc
FROM (
        SELECT
            product_id,
            SUM(CASE WHEN event_name = 'order' THEN 1 ELSE 0 END) AS nbr_order,
            SUM(CASE WHEN event_name = 'seen' THEN 1 ELSE 0 END) AS nbr_seen,
            SUM(CASE WHEN event_name = 'add to cart' THEN 1 ELSE 0 END) AS nbr_atc
        FROM df_caudz
        GROUP BY product_id
    )
GROUP BY product_id
"""
df_question_2B = sqldf(q)
df_question_2B
```

# Question 4: Parcours clients par jour et par utilisateur

```python
q="""
-- On agrège les données afin de calculer les indicateurs qui nous intéressent
SELECT 
    category_parcours_user,
    COUNT(user_id) AS times_category_appears,
    COUNT(DISTINCT user_id) AS nbr_different_user_id,
    SUM(total_cost) AS total_cost_for_category, 
    SUM(total_quantity) AS total_quantity_for_category
FROM
(
-- En fonction des actions réalisées et de leur ordre, on crée une catégorie dédiée
    SELECT
        user_id,
        parcours_user,
        total_cost, 
        total_quantity,
        CASE 
            WHEN seen_locate <> 0 AND atc_locate = 0 AND order_locate = 0 THEN 'seen'
            WHEN seen_locate = 0 AND atc_locate <> 0 AND order_locate = 0 THEN 'add to cart'
            WHEN seen_locate = 0 AND atc_locate = 0 AND order_locate <> 0 THEN 'order'

            WHEN seen_locate <> 0 AND atc_locate <> 0 AND order_locate = 0 THEN
                CASE
                    WHEN seen_locate < atc_locate THEN 'seen, atc'
                    ELSE 'atc, seen'
                END

            WHEN seen_locate <> 0 AND atc_locate = 0 AND order_locate <> 0 THEN
                CASE
                    WHEN seen_locate < order_locate THEN 'seen, order'
                    ELSE 'order, seen'
                END

            WHEN seen_locate = 0 AND atc_locate <> 0 AND order_locate <> 0 THEN
                CASE
                    WHEN atc_locate < order_locate THEN 'atc, order'
                    ELSE 'order, atc'
                END

            WHEN seen_locate <> 0 AND atc_locate <> 0 AND order_locate <> 0 THEN
                CASE
                    WHEN seen_locate < atc_locate AND atc_locate < order_locate THEN 'seen, atc, order'
                    WHEN seen_locate < order_locate AND order_locate < atc_locate THEN 'seen, order, atc'
                    WHEN atc_locate < seen_locate AND seen_locate < order_locate THEN 'atc, seen, order'
                    WHEN atc_locate < order_locate AND order_locate < seen_locate THEN 'atc, order, seen'
                    WHEN order_locate < seen_locate AND seen_locate < atc_locate THEN 'order, seen, atc'
                    WHEN order_locate < atc_locate AND atc_locate < seen_locate THEN 'order, atc, seen'
                END
        END AS category_parcours_user
    FROM 
    (
    -- On détermine quelles sont les actions qui ont été réalisées, et dans quel ordre
        SELECT
            user_id,
            parcours_user,
            total_cost, 
            total_quantity,
            INSTR(parcours_user, 'seen') AS seen_locate,
            INSTR(parcours_user, 'add to cart') AS atc_locate,
            INSTR(parcours_user, 'order') AS order_locate
        FROM
        (  
        -- On récupère les couts et quantités + on crée une chaine de caractères contenant toutes les actions réalisées pas un user_id DANS une journée
            SELECT 
                user_id,
                date,
                SUM(sales) AS total_cost,
                SUM(quantity) AS total_quantity,
                GROUP_CONCAT(event_name, ', ') AS parcours_user
            FROM 
            (
            -- On trie dans l'ordre des user_id et des dates pour avoir la bonne séquence d'actions
                SELECT *
                FROM df_caudz
                ORDER BY user_id, date 
            )
            GROUP BY date, user_id
            ORDER BY user_id
        )
    )
)
GROUP BY category_parcours_user
ORDER BY times_category_appears DESC
"""
sqldf(q)
```

# Question 5: Quels sont les user_id qui ont acheté au moins 2 produits différents la même journée

```python
q="""
SELECT
    user_id,
    date,
    SUM(sales*quantity) AS total_cost,
    COUNT(DISTINCT product_id) AS nbr_products_bought
FROM
(
    -- On récupère uniquement les données des 'order' et on les trie par user_id et par jour
    SELECT *
    FROM df_caudz
    WHERE event_name='order'
    ORDER BY user_id, date
)
GROUP BY user_id, date
HAVING nbr_products_bought >= 2
"""
sqldf(q)
```

# Question 6: En comparant avec la table df_lowcost :

### A: Quels sont les user_id qui ont acheté le même jour sur 2 tables différentes ?

```python
q="""
SELECT
    user_id, 
    date, 
    CASE
        WHEN INSTR(tables_achats, 'df_caudz')<> 0 THEN 'yes'
        ELSE 'no'
    END AS achat_df_caudz,
    CASE
        WHEN INSTR(tables_achats, 'df_lowcost') <> 0 THEN 'yes'
        ELSE 'no'
    END AS achat_df_lowcost
FROM
(
    -- On regroupe par user_id et par date, les tables ou l'utilisateur a réalisé un achat
    SELECT
        user_id,
        date,
        GROUP_CONCAT(table_origine, ', ') AS tables_achats
    FROM
    (
        SELECT
            user_id,
            table_origine,
            date,
            event_name,
            sales,
            quantity
        FROM
        (
            --Union des tables
            SELECT
                'df_caudz' AS table_origine,
                *
            FROM df_caudz

            UNION

            SELECT
                'df_lowcost' AS table_origine,
                *
            FROM df_lowcost
        )
        WHERE event_name = 'order'
    )
    GROUP BY user_id, date
)
WHERE achat_df_caudz = 'yes' AND achat_df_lowcost = 'yes'
"""
sqldf(q)
```

### B: Quels sont les user_id qui n'ont acheté que sur df_caudz et ceux que sur df_lowcost ?

```python
q="""
SELECT
    user_id, 
    date, 
    CASE
        WHEN INSTR(tables_achats, 'df_caudz')<> 0 THEN 'yes'
        ELSE 'no'
    END AS achat_df_caudz,
    CASE
        WHEN INSTR(tables_achats, 'df_lowcost') <> 0 THEN 'yes'
        ELSE 'no'
    END AS achat_df_lowcost
FROM
(
    -- On regroupe par user_id et par date, les tables ou l'utilisateur a réalisé un achat
    SELECT
        user_id,
        date,
        GROUP_CONCAT(table_origine, ', ') AS tables_achats
    FROM
    (
        SELECT
            user_id,
            table_origine,
            date,
            event_name,
            sales,
            quantity
        FROM
        (
            --Union des tables
            SELECT
                'df_caudz' AS table_origine,
                *
            FROM df_caudz

            UNION

            SELECT
                'df_lowcost' AS table_origine,
                *
            FROM df_lowcost
        )
        WHERE event_name = 'order'
    )
    GROUP BY user_id, date
)
WHERE achat_df_caudz <> achat_df_lowcost
"""
sqldf(q)
```
